### Constant

```python
a = tf.constant(3)
b = tf.constant(3,tf.float32)
c = tf.constant([2,5,6,9,6])
d = tf.constant([[2,5,6],
                 [8,5,8],
                 [8,8,1]])
e = tf.constant('Start with TensorFlow')
```

```python
with tf.Session() as sess : 
    print(sess.run(a))
    print(sess.run(b))
    print(sess.run(c))
    print(sess.run(d))
    print(sess.run(e))
```

### Variable

```python
a = tf.Variable(3)
b = tf.Variable(3,tf.float32)
c = tf.Variable([2,5,6,9,6])
d = tf.Variable([[2,5,6],[8,5,8], [8,8,1]])
e = tf.Variable('Start with TensorFlow')
```

```python
with tf.Session() as sess :
    sess.run(tf.global_variables_initializer())
    print(sess.run(a))
    print(sess.run(b))
    print(sess.run(c))
    print(sess.run(d))
    print(sess.run(e))
```

```python
price = tf.Variable(20) 
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
print('original price is : ',sess.run(price))
```

```python
price = tf.Variable(30) 
sess.run(tf.global_variables_initializer())
print('the price is : ',sess.run(price))

price = tf.Variable(40) 
sess.run(tf.global_variables_initializer())
print('the price is : ',sess.run(price))

price = tf.Variable(50) 
sess.run(tf.global_variables_initializer())
print('the price is : ',sess.run(price))

price = tf.Variable(60) 
sess.run(tf.global_variables_initializer())
print('the price is : ',sess.run(price))
```

```python
price = tf.assign(price,50)
print('new price is : ',sess.run(price))
```

```python
for j in range(6) : 
    price = tf.assign(price,price+10)
    print('new price is : ',sess.run(price))
```

### Placeholder

```python
a = tf.placeholder(tf.float32)
with tf.Session() as sess : 
    result = sess.run(a , feed_dict ={a:3})
    print(result)
```

```python
a = tf.placeholder(tf.float32)
b = a*2
with tf.Session() as sess : 
    result = sess.run(b , feed_dict ={a:3})
    print(result)
```

```python
x = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
z = tf.placeholder(tf.float32)
w = tf.placeholder(tf.float32)

b = x*2 + 3*y**4 - 4/z + w-9

with tf.Session() as sess : 
    result = sess.run(b , feed_dict ={x:3,y:4,z:8,w:6})
    print(result)
```

```python
x = tf.placeholder(tf.float32)
y = x**2
with tf.Session() as sess :
    print(sess.run(y,feed_dict={x:5}))
```

```python
x = tf.placeholder(tf.float32)
y = x**2
with tf.Session() as sess :
    print(sess.run(y, {x:[5,8,10,50]}))
```

```python
x = tf.placeholder(tf.float32)
y = x**2
with tf.Session() as sess :
    print(sess.run(y, {x:list(range(5,100))}))
```

```python
w = tf.constant([5.0])
b = tf.Variable([7.0],tf.float32)

x = tf.placeholder(tf.float32)

LinearRegressionModel = w*x + b

with tf.Session() as sess :
    sess.run(tf.global_variables_initializer())
    print(sess.run(LinearRegressionModel,{x:[3,5,7,9]}))
```

### Operations

```python
s1 = tf.constant('Hi, ')
s2 = tf.constant('people ')
s3 = tf.add(s1,s2)

sess = tf.Session()
print(sess.run(s3))
```

```python
matrix1 = tf.constant([[5., 3.]])
matrix2 = tf.constant([[2.],
                       [2.]])
product = tf.matmul(matrix1, matrix2)

with tf.Session() as sess:
    result = sess.run(product)
    print (result)
```

```python
matrix1 = tf.constant([[5,6,9,8],
                      [1,2,8,9],
                      [7,9,5,6],
                      [1,2,3,9]])
```

```python
matrix2 = tf.matrix_transpose(matrix1)

with tf.Session() as sess:
    result = sess.run(matrix2)
    print (result)
```
